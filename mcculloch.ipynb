```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
def activation_func(x,theta=1):
     return 1 if x>=theta else 0
```

```python
def McCulloch_Pitts(x,t,w):
    """
    x : list 
        Input .
    t : list
        Target .
    w : list
        Weights.
    b : int 
        Bias.
    """

    tPred = []
     
    for i in range(0,len(x)):
        yin = x[i][0]*w[0] + x[i][1]*w[1];
        tCap = activation_func(yin)
        tPred.append(tCap)
        print(f"Input: ({x[i][0]}, {x[i][0]}), Yin: {yin}, tCap: {tCap}, t: {t[i]}")

    
    return tPred
```

AND GATE

```python
x =[[0,0],[0,1],[1,0],[1,1]]
t = [0, 0, 0, 1]

w = [1, 1]  # Weights for x and y

t_pred = McCulloch_Pitts(x, t, w)
```

OR GATE

```python
# OR gate implementation
x = [[0,0],[0,1],[1,0],[1,1]]
t = [0, 1, 1, 1]

w = [1, 1]  # Weights for x and y

t_pred = McCulloch_Pitts(x, t, w)
```
