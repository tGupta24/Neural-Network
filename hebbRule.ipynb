```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
def activationFunction(x):
    return 1 if x >= 0 else -1
```

```python
def  HebbRule(x,t,w,b):
    """
    HebbRule

    
    x : list 
        Input .
    t : list
        Target .
    w : list
        Weights.
    b : int 
        Bias.
    """

    for i in range(len(x)):
      yIn = w[0]*x[i][0] + w[1]*x[i][1]
      tPred = activationFunction(yIn)

      w1new = w[0] + x[i][0]*t[i]
      w2new = w[1] + x[i][1]*t[i]
      bnew = b+ t[i]

      w = [w1new,w2new]
      b = bnew

      print(f'Updated weights: {w}, {b} t = {t[i]} tPred = {tPred}')

    return w,b
 
```

AND GATE

```python
w = [1 , 1]
b = 1
x = np.array([[0,0],[0,1],[1,0],[1,1]])

t = np.array([0, 0, 0, 1])  # Target
w, b = HebbRule(x, t, w, b)
print(f'Final Weights: {w}, Bias: {b}')
```

Polar AND Gate

```python
w = [1 , 1]
b = 1
x = np.array([[1,1],[1,-1],[-1,1],[-1,-1]])

t = np.array([1, -1, -1, -1])  # Target
w, b = HebbRule(x, t, w, b)
print(f'Final Weights: {w}, Bias: {b}')
```
